S_k(n) = e^{jk\frac{2\pi}{N}n}

3. 画出N = 4, 5, 6, 7的Re{S_k(n)}的波形

```python
import numpy as np

N_list = np.array([4, 5, 6, 7,8])
k_list = np.arange(1, 10)  # Adjusted to ensure k_list has valid values for subplot
fig = figure(figsize=(16,12))
for N in N_list:
    for k in k_list:
        x_plot = np.linspace(0, 8, 1000, endpoint=True)
        x_stem = np.linspace(0, 8, 17, endpoint=True)
        y_plot = np.cos(k * 2 * np.pi * x_plot / N)
        y_stem = np.cos(k * 2 * np.pi * x_stem / N)
        ax = fig.add_subplot(len(k_list), len(N_list), (k - 1) * len(N_list) + np.where(N_list == N)[0][0] + 1)
        if N == N_list.min() and k != k_list.max():
            ax.set_xticks([])
        if k != k_list.max() and N != N_list.min():
            ax.set_yticks([])
            ax.set_xticks([])
        if k == k_list.max() and N!= N_list.min():
            ax.set_yticks([])
        ax.plot(x_plot, y_plot, linewidth=1, ls = '--', color = 'c')
        ax.stem(x_stem, y_stem, markerfmt='r.', basefmt='k-')
        ax.set_title(f'N = {N}, k = {k}', fontsize = 10, y = 0.95)
        ax.set_ylim([-1.1, 1.1])
        
```

```python
f = 10
fig = figure(figsize=(8,15))
fs_list = np.array([10, 50, 50.1])
for fs in fs_list:
    x_plot = np.arange(0, 1, 0.001)
    y_plot = np.sin(2 * np.pi * f * x_plot)
    x_stem = np.arange(0, 1+1/fs, 1/fs)
    y_stem = np.sin(2 * np.pi * f * x_stem)
    ax = fig.add_subplot(len(fs_list),1, np.where(fs_list == fs)[0][0] + 1)
    ax.plot(x_plot,y_plot,ls = '--', color = 'c')
    ax.stem(x_stem,y_stem,basefmt='k-',markerfmt='r.')
    if fs!=fs_list.max():
        ax.set_xticks([])
```
