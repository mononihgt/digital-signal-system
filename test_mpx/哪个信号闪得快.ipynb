```python
from pylab import *
from sympy import *
from matplotlib_inline import backend_inline
backend_inline.set_matplotlib_formats('svg')
import warnings
warnings.filterwarnings("ignore")
rcParams['font.family'] ='DejaVu Sans'
```

```python

n = 10
T_s = 1/60
x_1 =  np.linspace(0, n, n*1000, endpoint=False) 
x = np.linspace(0,n,n*60,endpoint=False)
y_1 = sin(2*np.pi*x/20/T_s)
y_2 = sin(2*np.pi*61*x/40/T_s)
y_3 = y_1 + sin(2*np.pi*(x/T_s-1)/20)
```

```python
y = np.fft.fft(y_1)
y = np.abs(y)
plot(y)
```

```python
fig = plt.figure(figsize=(20, 9))
ax = fig.add_subplot(311)
ax.stem(x,y_1)
ax.plot(x_1,sin(2*np.pi*x_1/20/T_s),color='r')
ax.set_title('Signal A')
ax = fig.add_subplot(313)
ax.stem(x,y_2)
ax.plot(x_1,sin(2*np.pi*61*x_1/40/T_s),color='r')
ax.set_title('Signal B')
ax = fig.add_subplot(312)
ax.stem(x,y_3)
ax.plot(x_1,sin(2*np.pi*x_1/20/T_s)+sin(2*np.pi*(x_1/T_s-1)/20),color='r')
ax.set_title('Signal A + Signal B')
plt.show()
```

```python
fig = plt.figure(figsize=(20, 3))
ax = fig.add_subplot(111)
ax.stem(x,y_3)
ax.plot(x,y_3,color='r')
ax.set_title('Signal C')
```

```python
fs = 20e3                    # 
f  = 1e3                     # 
t  = np.arange(0,0.1,1/fs)   # 00.1=1/fs
x  = 0.32*np.sin(2*np.pi*f*t)# 
y = np.fft.fft(x)
freq  = np.linspace(0,fs,int(len(y)))
y_amp = abs(y)

figure(figsize=(40, 3))
plot(freq,y_amp)
show()
```

```python
x = symbols('x')
T_s = 1/60
y_1 = sin(2*np.pi*x/20/T_s)
y_2 = sin(2*np.pi*61*x/40/T_s)
y_3 = y_1 + sin(2*np.pi*(x/T_s-1)/20)
A = 2 * np.pi * x / (20 * T_s)
B = 2 * np.pi * (x / T_s - 1) / 20

#  (A+B)/2  (A-B)/2
sum_half = (A + B) / 2
diff_half = (A - B) / 2

# 
y_3_product = 2 * sin(sum_half) * cos(diff_half)

# 
y_3_product_simplified = simplify(y_3_product)

# 
print(" (y_3):")
print(y_3)
print("\n:")
print(y_3_product_simplified)
```
