```python
import pandas as pd
import os
import numpy as np

df = pd.read_csv(os.path.join(os.getcwd(),'..','glove.6B','glove.6B.50d.csv'), index_col = 0)
df.head()
```

```python
# 对df每一行进行RMSNorm，即对每一行的元素进行平方和开根号，然后除以平方和开根号的值
def rmsnorm(row):
    return row/np.sqrt(np.sum(row**2)/len(row))

df_norm = df.apply(rmsnorm, axis=1)
df_norm.to_csv(os.path.join(os.getcwd(),'..','glove.6B','glove.6B.50d_rmsnorm.csv'))
df_norm.head()
```

```python
# 计算“dog”和“bone”、“bird”之间的距离
print(f"dog and bone: {df_norm.loc['dog'].dot(df_norm.loc['bone'])}")
print(f"dog and bird: {df_norm.loc['dog'].dot(df_norm.loc['bird'])}")
```

```python
# 计算dog和其他词的内积，并取前面10个最大值（除了dog本身） 
dog_norm_dot = df_norm.dot(df_norm.loc['dog'])
print(dog_norm_dot.nlargest(11)[1:])
```
