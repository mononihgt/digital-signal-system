```python
from pylab import *
import scipy.io as scio
import warnings
warnings.filterwarnings("ignore")
rcParams['font.family'] = ['DejaVu Sans','Simhei']

# shuohua
ms_shuohua = scio.loadmat('materials/shuohua_ms.mat')['ms']
ms_rms_shuohua = scio.loadmat('materials/shuohua_ms_rms.mat')['ms_rms']
f_shuohua = scio.loadmat('materials/shuohua_f.mat')['f']
# tanshe
ms_tanshe = scio.loadmat('materials/tanshe_ms.mat')['ms']
ms_rms_tanshe = scio.loadmat('materials/tanshe_ms_rms.mat')['ms_rms']
f_tanshe = scio.loadmat('materials/tanshe_f.mat')['f']
# song
ms_song6 = scio.loadmat('materials/song6_ms.mat')['ms']
ms_rms_song6 = scio.loadmat('materials/song6_ms_rms.mat')['ms_rms']
f_song6 = scio.loadmat('materials/song6_f.mat')['f']
# dialect
ms_dialect = scio.loadmat('materials/dialect_ms.mat')['ms']
ms_rms_dialect = scio.loadmat('materials/dialect_ms_rms.mat')['ms_rms']
f_dialect = scio.loadmat('materials/dialect_f.mat')['f']
# cello
ms_cello = scio.loadmat('materials/cello_ms.mat')['ms']
ms_rms_cello = scio.loadmat('materials/cello_ms_rms.mat')['ms_rms']
f_cello = scio.loadmat('materials/cello_f.mat')['f']



fig = figure(figsize=(8, 6), dpi=200)
ax = fig.add_subplot(111)
# ax.plot(f_shuohua[:len(ms_shuohua)], ms_rms_shuohua, label='Mandarin', linewidth=1)
# ax.plot(f_tanshe[:len(ms_tanshe)], ms_rms_tanshe, label='tongue trill',linewidth=1)
# ax.plot(f_tanshe[:len(ms_dialect)], ms_rms_dialect, label='dialect',linewidth=1)
# ax.plot(f_tanshe[:len(ms_song6)], ms_rms_song6, label='song',linewidth=1)
ax.plot(f_cello[:len(ms_cello)], ms_rms_cello, label='cello',linewidth=1)
ax.set_xlim([0.5, 64])
ax.set_xscale('log')
ax.set_xticks([0.5, 1, 2, 4,8, 16, 32, 64])
ax.get_xaxis().set_major_formatter(matplotlib.ticker.FuncFormatter(lambda x, _: f'{x:.1f}'))
ax.set_xlabel('frequency (Hz)', fontsize=16)
ax.set_ylabel('normalized amplitude', fontsize=16)
ax.set_title('modulation spectrum', fontsize=16)
ymin, ymax = ax.get_ylim()
ax.set_ylim([ymin, ymax])
# ax.vlines([3.8333], -2, 0.9723, linestyles='dashed', colors='#cccccc', linewidth=0.8)
# ax.vlines([6], -2, 1, linestyles='dashed', colors='#999999', linewidth=0.8)
# ax.vlines([5.5], -2, 0.945329, linestyles='dashed', colors='#cccccc', linewidth=0.8)
ax.legend()
for label in ax.get_xticklabels():
    label.set_rotation(45)
# fig.savefig('manuscript/figure/spectrum_song.png', dpi=300, bbox_inches='tight')
# fig.savefig('manuscript/figure/spectrum_tanshe.png', dpi=300, bbox_inches='tight')
# fig.savefig('manuscript/figure/spectrum_dialect.png', dpi=300, bbox_inches='tight')
# fig.savefig('manuscript/figure/spectrum_shuohua.png', dpi=300, bbox_inches='tight')
show()
```
